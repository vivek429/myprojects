```python
#importing necssary library
import pandas as pd
import numpy as np 
import sklearn as sk
import nltk
nltk.download("wordnet")
```

```python
#importing countvectorizer from sklearn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
filepath_dict= {"yelp":"yelp_labelled.txt","amazon":"amazon_cells_labelled.txt","imdb":"imdb_labelled.txt"}# creating dictionary key as source and value as path of file
```

```python
#data set has been taken from uci machine learning repository
#this is a link
#https://archive.ics.uci.edu/ml/datasets/Sentiment+Labelled+Sentences
# creating data frame for each file andd appending them into a list
df_list=[]
for source,filepath in filepath_dict.items():
    df=pd.read_csv(filepath,names=["sentence","lable"],sep="\t")
    df["source"]=source
    df_list.append(df)
```

```python
df_list
```

```python
df=pd.concat(df_list) #merging all the files in one and creating dataframe
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
vect=CountVectorizer() # creating an intance of countvetorizer
```

```python
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
# creating a class to use lemmatizer with tokenizer so tha we can use it with countvectorizer
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, articles):
        return [self.wnl.lemmatize(t) for t in word_tokenize(articles)]
```

```python
vect.set_params(tokenizer=LemmaTokenizer())# using lemmatokenizer with countvectorizer
```

```python
# removing english stop words
vect.set_params(stop_words="english")
```

```python
# including 1-gram and 2-gram
vect.set_params(ngram_range=(1,2))# this is for bigram
                                  #ngram_range= (1,1) for monogram and
                                  #ngram_range=(1,3) for trigram
```

```python
#ignore terms that appear in more than 50% of documents
vect.set_params(max_df=0.5)
```

```python

vect.set_params(min_df=0)
```

```python
x= d.drop("lable",axis=1)# independent variable
y=d["lable"]#target variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     x, y, test_size=0.3, random_state=100)#splitting the data into training and testing
```

```python
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# this is taking soo much tym so better skip it
folds = KFold(n_splits = 5, shuffle = True, random_state = 101)

# specify range of hyperparameters
# Set the parameters by cross-validation
hyper_params = [ {'gamma': [1e-2, 1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]}]


# specify model
model = SVC(kernel="rbf")

# set up GridSearchCV()
model_cv = GridSearchCV(estimator = model, 
                        param_grid = hyper_params, 
                        scoring= 'accuracy', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True )      

# fit the model
model_cv.fit(X_train, y_train)                  
```

```python
# non-linear model
# using rbf kernel, C=1, default value of gamma


# model
non_linear_model = SVC(kernel='rbf')

# fit
non_linear_model.fit(X_train, y_train)

# predict
y_pred = non_linear_model.predict(X_test)
```

```python
# confusion matrix and accuracy

# accuracy
print("accuracy:", metrics.accuracy_score(y_true=y_test, y_pred=y_pred), "\n")

# cm
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred))
```

```python
# linear model

model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)

# predict
y_pred = model_linear.predict(X_test)
```

```python
# confusion matrix and accuracy

# accuracy
print("accuracy:", metrics.accuracy_score(y_true=y_test, y_pred=y_pred), "\n")

# cm
print(metrics.confusion_matrix(y_true=y_test, y_pred=y_pred))
```
